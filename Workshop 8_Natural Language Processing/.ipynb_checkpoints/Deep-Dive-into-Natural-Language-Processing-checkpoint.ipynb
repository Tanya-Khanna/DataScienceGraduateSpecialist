# What is Sentiment Analysis?

Sentiment Analysis is a technique in Natural Language Processing (NLP) that identifies and categorizes opinions expressed in a piece of text, especially in order to determine whether the writer's attitude towards a particular topic, product, etc., is positive, negative, or neutral. Applications of sentiment analysis include monitoring brand and product sentiment in customer feedback, and understanding customer needs.

# What is Topic Modeling?

Topic Modeling is an NLP technique used to uncover abstract topics within a text. It involves clustering groups of similar words together in topics. Unlike sentiment analysis, which seeks to understand the *opinion* expressed, topic modeling aims to uncover the *subjects* being discussed. This is particularly useful for browsing, searching, or organizing large archives of texts by topic.

## Step 3: Understanding Sentiment Analysis with TextBlob

TextBlob is a Python library designed for processing textual data, offering a straightforward API for tackling common natural language processing (NLP) tasks. Among its capabilities, sentiment analysis is a prominent feature, providing insights into the emotional tone behind words.

### Sentiment Analysis Overview

Sentiment analysis in TextBlob yields two key metrics: polarity and subjectivity.

- **Polarity**: This metric falls within the range [-1.0, 1.0], indicating the sentiment direction. A score of -1.0 represents a completely negative sentiment, 0 stands for neutral, and 1.0 signifies entirely positive sentiment. This measure is derived from the average sentiment scores of individual words in the text, where each word's sentiment score contributes to the overall sentiment of the entire piece.

- **Subjectivity**: Ranging from 0.0 to 1.0, subjectivity quantifies the amount of personal opinion and factual information contained in the text. A score of 0.0 is very objective (fact-based), and 1.0 is very subjective (opinion-based). Like polarity, subjectivity scores for the text are computed based on the subjectivity scores of the individual words.

### How TextBlob Performs Sentiment Analysis

TextBlob leverages the pattern library's sentiment module, which includes a comprehensive list of words along with their associated positive or negative sentiment scores and subjectivity scores. The process is as follows:

1. The input text is tokenized into words and phrases.
2. Each token (word or phrase) is matched against the pattern library's sentiment lexicon to assign sentiment scores.
3. These scores are aggregated to produce an overall sentiment score for the text, reflecting the average sentiment orientation and subjectivity level of the text as a whole.

```python
### Example Code Snippet

from textblob import TextBlob

text = "I love this product. It works wonderfully."
blob = TextBlob(text)

print(blob.sentiment)
```

In this code snippet, positive words like "love" and "wonderfully" influence the overall polarity score, resulting in a high polarity score indicating positive sentiment. The computation of the subjectivity score similarly depends on the subjectivity levels of the words in the text.

### Limitations

While TextBlob's sentiment analysis offers a quick method for gauging the sentiment of text, it operates on a relatively simple level, primarily relying on a lexical approach that may not fully capture contextual nuances, sarcasm, or complex expressions. For more sophisticated analyses, especially those requiring an understanding of context or the detection of subtle emotional cues, advanced models based on deep learning techniques might be more appropriate.

```python
# Define a function to compute sentiment polarity using TextBlob
def get_sentiment(text):
    """
    This function takes a text as input and returns the sentiment polarity.
    Polarity score ranges from -1 (most negative) to 1 (most positive).
    """
    return TextBlob(text).sentiment.polarity

# Apply the function to each tweet and create a new column 'Sentiment'
df['Sentiment'] = df['Text'].apply(get_sentiment)
```

```python
# Defining a function to categorize sentiment based on polarity score
def categorize_sentiment(polarity):
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
df['Sentiment_Category'] = df['Sentiment'].apply(categorize_sentiment)
```

```python
df.head(10)
```

```python
# Define a function for basic text preprocessing
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Tokenize the text into words
    tokens = word_tokenize(text)
    # Remove stopwords
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    # Remove punctuation and numbers
    tokens = [word for word in tokens if word.isalpha()]
    return tokens
```

```python
# Apply text preprocessing to the 'Text' column
df['Processed_Text'] = df['Text'].apply(preprocess_text)
```

```python
df.head(10)
```

```python
from collections import Counter

# Function to plot top N words for a given sentiment category
def plot_top_words(sentiment, N=10):
    """Plot the top N words for tweets in a given sentiment category using seaborn."""
    # Filter the DataFrame for the given sentiment
    sentiment_df = df[df['Sentiment_Category'] == sentiment]
    
    # Flatten list of processed text into a single list and count words
    all_words = [word for text_list in sentiment_df['Processed_Text'] for word in text_list]
    word_counts = Counter(all_words)
    
    # Extract most common words and their counts
    common_words = word_counts.most_common(N)
    words, counts = zip(*common_words)
    
    # Convert to DataFrame for easy plotting with seaborn
    data = pd.DataFrame({'Words': words, 'Frequency': counts})
    
    # Creating the bar plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Frequency', y='Words', data=data, palette='viridis')
    plt.title(f'Top {N} Words in {sentiment} Sentiment Tweets', fontsize=14)
    plt.xlabel('Frequency', fontsize=12)
    plt.ylabel('Words', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.show()

# Plotting top words for each sentiment category
for sentiment in ['Positive', 'Negative', 'Neutral']:
    plot_top_words(sentiment, N=10)
```

## Step 4: Understanding Topic Modeling with LDA and Gensim

Topic Modeling is a method in natural language processing (NLP) to uncover hidden thematic structures in a large collection of texts. It enables the identification of topics present in a corpus, providing insights into the dominant themes without reading through all the documents. One of the most popular topic modeling techniques is Latent Dirichlet Allocation (LDA).

### Latent Dirichlet Allocation (LDA)

LDA is a generative statistical model that allows sets of observations to be explained by unobserved groups that explain why some parts of the data are similar. In the context of text data, these "observations" are words collected into documents, and it assumes that:

- Each document can be described as a distribution over topics.
- Each topic can be described as a distribution over words.

This approach to topic modeling can unearth the underlying topic structure of a corpus, making it easier to understand the main themes contained within it.

### Gensim for LDA

Gensim is a Python library designed to automatically extract semantic topics from documents, as efficiently (memory and CPU-wise) as possible. Gensim's LDA implementation is highly optimized and is capable of handling large text corpora.

### Process Overview

1. **Dictionary Creation**: Map each word in the corpus to a unique ID using Gensim's `corpora.Dictionary`.
2. **Corpus Creation**: Convert the list of documents (words) into a `Bag-of-Words` format - a list of (word ID, word frequency) tuples.
3. **Applying LDA**: Run LDA using the `gensim.models.LdaModel` on the corpus, specifying the number of topics.
4. **Reviewing Topics**: Analyze the top words for each topic to understand their themes.
5. **Visualization**: Use `pyLDAvis` to provide an interactive visualization, helping to interpret the topics.

### Why Topic Modeling?

Topic modeling is invaluable for summarizing a large corpus of text, discovering hidden thematic structures, and organizing the collection in a way that's interpretable. It's used extensively in digital humanities, content recommendation systems, and for understanding user feedback in social media, reviews, and more.

```python
# Importing necessary libraries
from gensim import corpora, models
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

# Step 1: Dictionary Creation
dictionary = corpora.Dictionary(df['Processed_Text'])

# Step 2: Corpus Creation
corpus = [dictionary.doc2bow(text) for text in df['Processed_Text']]

# Step 3: LDA Model Training
lda_model = models.LdaModel(corpus, num_topics=5, id2word=dictionary, passes=15, random_state=42)

# Display the topics
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```

# **Text Summarization with Natural Language Processing**

## **Introduction**

###### *In an era of information overload, where digital media and publishing are continually surging, who has the time to wade through entire articles, documents, or books to determine their utility? That's where automated text summarization steps in, offering a faster way to understand the essential ideas of expansive writings.*




Navigating through the sea of information, one often encounters superfluous details that cloud the main message. Summarization offers a solution by condensing this information into digestible chunks, eliminating redundancy, and highlighting core messages. Employing summarization techniques can boost document clarity, trim the duration of information search, and allow a denser presentation of data within a specific area. 

### **Problem Formulation**

##### Given:

- $f$ as the summarization function.
- $D$ as a document or set of documents.
- $q$ as a query.

The summarization task can be mathematically represented as:

1. **Single Document Summarization**:
   $summary = f(D)$

2. **Multi-document Summarization**:
   $summary = f(D_1, D_2, \dots, D_n)$

3. **Query-focused Summarization**:
   $summary = f(D, q)$

Where the possible outputs or types of "summary" can be:

- **Indicative Summary**: Gives a brief overview.
- **Informative Summary**: Provides full details.
- **Keyword Summary**: Lists key phrases.
- **Headline Summary**: One-line summary.

## **Extractive Summarization**

Extractive summarization is a text summarization technique that selects and extracts sentences or phrases directly from the source text to create a summary.
It aims to preserve the most important information by identifying and including key sentences without generating new content. 

*Input document → sentences similarity → weight sentences → select sentences with higher rank.*

If method X produces summaries that consist of sentences or phrases exactly as they appear in the original text, it is an extractive method. The sentences in the output will typically be found in the same form within the source document.

Pros:

*  Preserves original context and wording.
*  Simple and computationally efficient.

Cons:

*  May result in incoherent summaries.
*  Doesn't generate new insights or paraphrased content.
*  Limited to the sentences present in the source text.

### **Extractive Text Summarization using TF-IDF**

Before diving into the code, it's essential to grasp the concept of TF-IDF. TF-IDF (Term Frequency-Inverse Document Frequency) is a statistical measure used to evaluate how important a word is to a document in a collection or corpus. It's a fundamental concept in information retrieval, text mining, and NLP tasks.

### Components of TF-IDF:

#### 1. Term Frequency (TF):

This is a measure of the frequency of a term within a specific document. If a word is frequent in a document but not across many documents, it's relevant.

$$ \text{TF}(t,d) = \frac{\text{Number of times term } t \text{ appears in document } d}{\text{Total number of terms in document } d} $$

Where:
- $ t $ is the term or word.
- $ d $ is the document.

#### 2. Inverse Document Frequency (IDF):

This measure reflects the importance of a term. Reduces the weight for commonly used terms and increases the weight for terms that are not used in many documents. Common terms like "is", "of", and "that" might appear often but don't carry as much meaning.

$$ \text{IDF}(t, D) = \log\left(\frac{\text{Total number of documents in corpus } D}{\text{Number of documents with term } t \text{ in them}}\right) $$

Where:
- $ D $ is the corpus of documents.
- $ t $ is the term or word.

#### 3. TF-IDF:

Combines both the local and global insights. Words frequent in one document but not in many will have a high score. The TF-IDF score is simply the product of TF and IDF:

$$ \text{TF-IDF}(t, d, D) = \text{TF}(t,d) \times \text{IDF}(t, D) $$

Source for the following Dataset: https://owl.purdue.edu/owl/research_and_citation/using_research/quoting_paraphrasing_and_summarizing/paraphrasing_sample_essay.html

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sample Input Text
input_text = """
A decline in standardized test scores is but the most recent indicator that American education is in trouble. One reason for the crisis is that present mandatory-attendance laws force many to attend school who have no wish to be there. Such children have little desire to learn and are so antagonistic to school that neither they nor more highly motivated students receive the quality education that is the birthright of every American.
The solution to this problem is simple: Abolish compulsory-attendance laws and allow only those who are committed to getting an education to attend. This will not end public education. Contrary to conventional belief, legislators enacted compulsory-attendance laws to legalize what already existed. William Landes and Lewis Solomon, economists, found little evidence that mandatory-attendance laws increased the number of children in school. They found, too, that school systems have never effectively enforced such laws, usually because of the expense involved.
There is no contradiction between the assertion that compulsory attendance has had little effect on the number of children attending school and the argument that repeal would be a positive step toward improving education. Most parents want a high school education for their children. Unfortunately, compulsory attendance hampers the ability of public school officials to enforce legitimate educational and disciplinary policies and thereby make the education a good one.
Private schools have no such problem. They can fail or dismiss students, knowing such students can attend public school. Without compulsory attendance, public schools would be freer to oust students whose academic or personal behavior undermines the educational mission of the institution.
Has not the noble experiment of a formal education for everyone failed? While we pay homage to the homily, "You can lead a horse to water but you can't make him drink," we have pretended it is not true in education.
Ask high school teachers if recalcitrant students learn anything of value. Ask teachers if these students do any homework. Quite the contrary, these students know they will be passed from grade to grade until they are old enough to quit or until, as is more likely, they receive a high school diploma. At the point when students could legally quit, most choose to remain since they know they are likely to be allowed to graduate whether they do acceptable work or not.
Abolition of archaic attendance laws would produce enormous dividends.
First, it would alert everyone that school is a serious place where one goes to learn. Schools are neither day-care centers nor indoor street corners. Young people who resist learning should stay away; indeed, an end to compulsory schooling would require them to stay away.
Second, students opposed to learning would not be able to pollute the educational atmosphere for those who want to learn. Teachers could stop policing recalcitrant students and start educating.
Third, grades would show what they are supposed to: how well a student is learning. Parents could again read report cards and know if their children were making progress.
Fourth, public esteem for schools would increase. People would stop regarding them as way stations for adolescents and start thinking of them as institutions for educating America's youth.
Fifth, elementary schools would change because students would find out early they had better learn something or risk flunking out later. Elementary teachers would no longer have to pass their failures on to junior high and high school.
Sixth, the cost of enforcing compulsory education would be eliminated. Despite enforcement efforts, nearly 15 percent of the school-age children in our largest cities are almost permanently absent from school.
Communities could use these savings to support institutions to deal with young people not in school. If, in the long run, these institutions prove more costly, at least we would not confuse their mission with that of schools.
Schools should be for education. At present, they are only tangentially so. They have attempted to serve an all-encompassing social function, trying to be all things to all people. In the process they have failed miserably at what they were originally formed to accomplish.
"""
```

```python
# Function to tokenize a text into words, convert them to lowercase, and remove punctuation

def preprocess_text(text):
    # Tokenize text into sentences
    sentences = sent_tokenize(text)

    # Tokenize sentences into words, remove punctuation, and convert to lowercase
    processed_sentences = []
    for sentence in sentences:
        words = word_tokenize(sentence)
        words = [word.lower() for word in words if word.isalpha()]
        processed_sentences.append(' '.join(words))

    return processed_sentences
```

```python
# TF-IDF Vectorization

def tfidf_vectorization(text):
    # Preprocess the input text
    sentences = preprocess_text(text)

    # Create a TF-IDF vectorizer
    tfidf_vectorizer = TfidfVectorizer()

    # Fit and transform the sentences
    tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)

    return tfidf_vectorizer, tfidf_matrix
```

```python
# Calculate Cosine Similarity
def calculate_cosine_similarity(tfidf_matrix):
    # Calculate cosine similarity between sentences
    similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

    return similarity_matrix
```

```python
# Extractive Summarization using Cosine Similarity
def extractive_summarization(text, num_sentences=3):
    # Calculate TF-IDF matrix
    tfidf_vectorizer, tfidf_matrix = tfidf_vectorization(text)

    # Calculate cosine similarity matrix
    similarity_matrix = calculate_cosine_similarity(tfidf_matrix)

    # Get sentence scores based on similarity
    sentence_scores = similarity_matrix.sum(axis=1)

    # Get the indices of the top N sentences
    top_sentence_indices = sentence_scores.argsort()[-num_sentences:][::-1]

    # Sort the indices to maintain the original order
    top_sentence_indices.sort()

    # Generate the extractive summary
    extractive_summary = ""
    for index in top_sentence_indices:
        extractive_summary += preprocess_text(text)[index] + ' '

    return extractive_summary
```

```python
# Generate Extractive Summary
summary = extractive_summarization(input_text, num_sentences=3)
print("Extractive Summary:")
print(summary)
```
